```python
dots = sns.load_dataset('dots')
dots.head()
```

```python
dots = dots.query("align=='dots'")
dots
```

```python
sns.relplot(data=dots, kind='line', x='time', y='firing_rate', hue='coherence', style='choice');
```

```python
# palette 
palette = sns.cubehelix_palette(light=.8, n_colors=6)
sns.relplot(data=dots, kind='line', x='time', y='firing_rate', hue='coherence', style='choice',
           palette=palette);
```

```python
# logNorm , query 
from matplotlib.colors import LogNorm
sns.relplot(data=dots.query('coherence > 0'), kind='line', x='time', y='firing_rate',
           hue='coherence', hue_norm=LogNorm(),
           style='choice');
```

```python
# palette 
# sns.relplot(data=dots, kind='line', x='time', y='firing_rate', hue='coherence', size='choice');
sns.relplot(x='time', y='firing_rate',
           hue='coherence', size='choice',
           palette=palette,
           kind='line', data=dots);
```
