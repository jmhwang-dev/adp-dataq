```python
import seaborn as sns
sns.set_theme(style='darkgrid')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.time.unique()
```

```python
g = sns.FacetGrid(tips, col='time')
g.map(sns.histplot, "tip");
```

```python
sns.displot(x='tip', data=tips, col='time');
```

```python
g = sns.FacetGrid(data=tips, col='sex', hue='smoker')
g.map(sns.scatterplot, 'total_bill', 'tip', alpha=.7)
g.add_legend();
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', col='sex', hue='smoker',
           kind='scatter');
```

```python
g = sns.FacetGrid(tips, row="smoker", col="time", margin_titles=True)
g.map(sns.regplot, "size", "total_bill", color=".3", fit_reg=False, x_jitter=.0, y_jitter=.0);
```

```python
g = sns.FacetGrid(tips, col="day", height=4, aspect=.5)
g.map(sns.barplot, "sex", "total_bill", order=["Male", "Female"])
```

```python
g = sns.FacetGrid(tips, col="day", height=8, aspect=5)
g.map(sns.barplot, "sex", "total_bill", order=["Male", "Female"])
```

```python
g = sns.FacetGrid(tips, col="day", height=8, aspect=1.)
g.map(sns.barplot, "sex", "total_bill", order=["Male", "Female"])
```

```python
attend = sns.load_dataset('attention').query("subject <= 12")
attend.head()
```

```python
g = sns.FacetGrid(attend, col='subject', col_wrap=4, height=2, ylim=(0, 10))
g.map(sns.pointplot, 'solutions', 'score', order=[1,2,3], color='.3', ci=None)
```
