```python
import seaborn as sns
import pandas as pd
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.catplot(kind='count', data=titanic, x='sex', hue='survived');
```

```python
sns.catplot(kind='boxen', data=titanic, x='sex', y='age', hue='survived');
```

```python
sns.catplot(kind='boxen', data=titanic, x='sex', y='age', hue='survived');
```

```python
sns.relplot(
    kind='scatter', data=titanic, x='age', y='fare', hue='survived',
    col='sex', row='embarked', height=2, aspect=2);
```

```python
sns.catplot(kind='count', data=titanic, x='embarked', hue='survived');
```

```python
sns.catplot(kind='count', data=titanic, x='embark_town', hue='survived');
```

```python
sns.catplot(kind='count', data=titanic, x='who', hue='survived',
            col='embark_town', row='class');
```

```python
sns.catplot(kind='count', data=titanic, x='sex', hue='survived');
```

```python
sns.displot(kind='hist', data=titanic, x='age', hue='survived', col='sex');
```

```python
sns.displot(kind='hist', data=titanic, x='age', hue='survived', col='sex', row='class');
```
