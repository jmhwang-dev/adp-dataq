```python
import seaborn as sns
# sns.set_theme(style='darkgrid')
sns.set_theme(style="ticks", color_codes=True)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.catplot(data=tips, x='day', y='total_bill');
```

```python
sns.catplot(data=tips, x='day', y='total_bill', jitter=False);
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='swarm');
```

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='swarm');
```

```python
sns.catplot(data=tips, x='smoker', y='tip', order=['No', 'Yes']);
```

# Distributions of observations within categories

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='box');
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='box', hue='smoker');
```

```python
tips['weekend'] = tips['day'].isin(['Sat', 'Sun'])
sns.catplot(data=tips, x='day', y='total_bill', hue='weekend', kind='box');
```

```python
tips['weekend'] = tips['day'].isin(['Sat', 'Sun'])
sns.catplot(data=tips, x='day', y='total_bill', hue='weekend', kind='box', dodge=False);
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds
```

```python
sns.catplot(data=diamonds.sort_values('color'), x='color', y='price', kind='boxen');
```

```python
# violinplot: combine a boxplot with the kernel density estimation procedure described in the distbutions
sns.catplot(data=tips, x='total_bill', y='day', hue='sex', kind='violin');
```

```python
sns.catplot(data=tips, x='total_bill', y='day', hue='sex', kind='violin', bw=.15, cut=0);
```

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='violin', split=True);
```

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='violin',
            split=True, inner='stick', palette='pastel');
```

```python
g = sns.catplot(data=tips, x='day', y='total_bill', kind='violin', inner=None)
sns.swarmplot(data=tips, x='day', y='total_bill', color='k', size=3, ax=g.ax);
```

# Statistical estimation within categories

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.catplot(data=titanic, x='sex', y='survived', kind='bar', hue='class');
```

```python
sns.catplot(data=titanic, x='deck', kind='count', palette='ch:.25');
```

```python
sns.catplot(data=titanic, y='deck', hue='class', kind='count', palette='pastel', edgecolor=".6");
```

```python
sns.catplot(data=titanic, x='sex', y='survived', kind='point', hue='class');
```

```python
sns.catplot(data=titanic, x='class', y='survived', hue='sex', kind='point',
           palette={'male': 'g', 'female': 'm'},
           markers=["^", 'o'],
           linestyles=['-', '--']);
```

# Showing multiple relationships with facets

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='smoker', kind='swarm', col='time', aspect=.7);
```

```python
g = sns.catplot(data=titanic.query("fare > 0"), x='fare', y='survived',
                kind='box', row='class', orient='h', 
                height=1.5, aspect=4)
g.set(xscale='log');
```
